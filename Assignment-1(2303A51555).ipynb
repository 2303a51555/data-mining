```python
numeric_cols = ['Age', 'Fare', 'Parch', 'SibSp']
df2_numeric = df2[numeric_cols]

minmax_scaler = MinMaxScaler()
df2_minmax = pd.DataFrame(minmax_scaler.fit_transform(df2_numeric), columns=df2_numeric.columns)
print("Min-Max Normalized Data:\n", df2_minmax)
```

```python
zscore_scaler = StandardScaler()
df2_zscore = pd.DataFrame(zscore_scaler.fit_transform(df2_numeric), columns=df2_numeric.columns)

print("\nZ-score Standardized Data:\n", df2_zscore)
```
